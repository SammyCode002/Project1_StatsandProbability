```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_excel('Cleaned_Technology_Data.xlsx')

sns.set_theme(context='talk', style='whitegrid', palette='pastel', font='sans-serif', font_scale=1.1)

palette = sns.color_palette("coolwarm", 15)

# histogram for the 'Hours Using' column
plt.figure(figsize=(12, 7))
sns.histplot(data['Hours Using'], bins=15, kde=True, color=palette[4])
plt.title('Distribution of Hours Using Technology Devices', fontsize=20)
plt.xlabel('Hours Using', fontsize=16)
plt.ylabel('Frequency', fontsize=16)

# Customize the grid and axes styles
plt.grid(True, linestyle='--', linewidth=0.6)
sns.despine(trim=True, offset=10)

plt.show()

```

```python
# box plot for the 'Hours Using' column to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Hours Using'])
plt.title('Box Plot of Hours Using Technology Devices')
plt.xlabel('Hours Using')
plt.show()
```

```python
# line graph
pivot_data = data.pivot_table(index='Year', columns='Device', values='Hours Using', aggfunc='mean')

# line graph with the legend outside the plot area
plt.figure(figsize=(14, 8))
sns.lineplot(data=pivot_data, dashes=False, palette='tab10', linewidth=2.5)
plt.title('Trend of Hours Using Different Technology Devices Over Years')
plt.xlabel('Year')
plt.ylabel('Average Hours Using')
plt.legend(title='Device', title_fontsize='13', fontsize='11', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
